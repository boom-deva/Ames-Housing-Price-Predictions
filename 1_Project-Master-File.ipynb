### Imputing Missing Values: Ordinal Features

As discussed earlier, we should overwrite the ordinal features (i.e. categorical variables with values that have an implied order to them) with numerics to better fit the regression, so we've defined a dictionary mapping to replace it with numbers.

Based on the data dictionary, this is the list of ordinal features:

```python
ordinals = ['Street',       'Alley',         'Utilities',     'Land Slope',
            'Exter Qual',   'Exter Cond',    'Bsmt Qual',     'Bsmt Cond',
            'Bsmt Exposure','BsmtFin Type 1','BsmtFin Type 2','Heating QC',
            'Kitchen Qual', 'Functional',    'Fireplace Qu',  'Garage Finish',
            'Garage Qual',  'Garage Cond',   'Paved Drive',   'Pool QC']
```

Now we define dictionary mappings to replace ratings with numbers

```python
# Define Mappings
rating_5scale     = {'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,np.nan: 0}
rating_4scale     = {'Ex': 4,'Gd': 3,'TA': 2,'Fa': 1,np.nan: 0}
exposure_rating   = {'Gd': 4,'Av': 3,'Mn': 2,'No': 1,np.nan: 0}
finish_rating     = {'GLQ': 6,'ALQ': 5,'BLQ': 4,'Rec': 3,'LwQ': 2,'Unf': 1,np.nan: 0}
functional_rating = {'Typ': 8,'Min1': 7,'Min2': 6,'Mod': 5,'Maj1': 4,'Maj2': 3,'Sev': 2,'Sal': 1,np.nan: 0}
pave_rating       = {"Pave": 2, "Grvl": 1}
utils_rating      = {"AllPub": 4, "NoSewr": 3, "NoSeWa": 2, "ELO": 1}
slope_rating      = {"Gtl": 3, "Mod": 2, "Sev": 1}
garage_fin_rating = {"Fin": 3, "Rfn": 2, "Unf": 1, np.nan: 0}
drive_pave_rating = {"Y": 2, "P": 1, "N": 0, np.nan: 0}
```

We've created a function to apply to the data frame since we expect to apply this to the Kaggle test data.

```python
def clean_ordinals(data):
    data['Street'].replace(pave_rating, inplace=True)
    data['Alley'].replace(pave_rating, inplace=True)
    data['Utilities'].replace(utils_rating, inplace=True)
    data['Land Slope'].replace(slope_rating, inplace=True)
    data['Exter Qual'].replace(rating_5scale, inplace=True)
    data['Exter Cond'].replace(rating_5scale, inplace=True)
    data['Bsmt Qual'].replace(rating_5scale, inplace=True)
    data['Bsmt Cond'].replace(rating_5scale, inplace=True)
    data['Bsmt Exposure'].replace(exposure_rating, inplace=True)
    data['BsmtFin Type 1'].replace(finish_rating, inplace=True)
    data['BsmtFin Type 2'].replace(finish_rating, inplace=True)
    data['Heating QC'].replace(rating_5scale, inplace=True)
    data['Kitchen Qual'].replace(rating_5scale, inplace=True)
    data['Functional'].replace(functional_rating, inplace=True)
    data['Fireplace Qu'].replace(rating_5scale, inplace=True)
    data['Garage Finish'].replace(garage_fin_rating, inplace=True)
    data['Garage Qual'].replace(rating_5scale, inplace=True)
    data['Garage Cond'].replace(rating_5scale, inplace=True)
    data['Paved Drive'].replace(drive_pave_rating, inplace=True)
    data['Pool QC'].replace(rating_4scale, inplace=True)
    return(data)
```

## Feature Engineering

### Dummy Variable Creation

Nominal variables where values are strings cannot be readily processed by a regression model unless we can quantify them in some way

**One-Hot Encoding** is one such process to turn these into dummy variables while making sure to drop the first case to avoid the Dummy Variable Trap.

```python
# One-hot encode categorical variables:
train_dummies = pd.get_dummies(train_nominals, drop_first=True)
test_dummies = pd.get_dummies(test_nominals, drop_first=True)
```

We'll need to make sure that the number of dummy variable cases in both the Training and Testing data are the same, otherwise the same model can't be applied to both sets:

```python
if(train_dummies.shape[1] != test_dummies.shape[1]):
    # Get the exclusive dummies from each set
    train_only_dummies = set(train_dummies.columns) - set(test_dummies.columns)
    test_only_dummies  = set(test_dummies.columns) - set(train_dummies.columns)
    
    # Append the missing dummies to the dummies dataframe
    for test_dummy in test_only_dummies:
        train_dummies[test_dummy] = 0
    for train_dummy in train_only_dummies:
        test_dummies[train_dummy] = 0
        
    # Maintain same column order between two dataframes
    test_dummies = test_dummies[train_dummies.columns]
```

### Automatic Log Feature Selection

Based on our exploratory data analysis on univariate distributions, we realize the need to apply log transformations to numeric features that exhibit skewed distributions (similar to `SalePrice`) to cushion the impact of any inherent outliers.

We begin by screening for features that have skewed distributions by defining a rule:
***moderately skewed distributions with an absolute skewness measure of $>$ 0.5 should be logged***

```python
# Compute skewness for all numeric columns based on TRAIN
skewness = train_numerics.apply(lambda x: stats.skew(x))

# Filter columns that exhibit |Skew| > 0.5
skewness = skewness[abs(skewness) > 0.5]

# Exclude ordinals since it doesn't make sense to apply logarithms to non-continuous values
skewed_features = skewness.index
skewed_features = list(set(skewed_features) - set(ordinals))

# Overwrite feature in original dataframe with the corresponding logged version
train_numerics[skewed_features] = np.log1p(train_numerics[skewed_features].copy())
test_numerics[skewed_features]  = np.log1p(test_numerics[skewed_features].copy())

print("We have",skewness.shape[0],"skewed numeric features that we are going to log transform")
print("These were:",skewed_features)
```

Now we save the data with all our newly engineered features.

```python
# Create master TRAIN data with all original features and our engineered features
master_train_data = train_numerics.join(train_dummies).join(y)
master_train_data.to_csv("./datasets/backup_files/train_master_model_data.csv")
print("Master Train Data has",master_train_data.shape[1],"columns")

# Create master TEST data with all original features and our engineered features
master_test_data = test_numerics.join(test_dummies).join(y)
master_test_data.to_csv("./datasets/backup_files/test_master_model_data.csv")
print("Master Test Data has",master_test_data.shape[1],"columns")
```

## Modeling

*Before we go ahead, it will be helpful to have some helper functions handy*

```python
# Write function to automate graphing process
def plot_predictions(model_name, y_train_pred, y_test_pred, y_train_actual, y_test_actual):
    plt.scatter(y_train_pred, y_train_actual, c = "blue", marker = "s", label = "Training data")
    plt.scatter(y_test_pred, y_test_actual, c = "lightgreen", marker = "s", label = "Testing data")
    plt.title(model_name+"Regression")
    plt.xlabel("$\hat{y}$ = Predicted SalePrice")
    plt.ylabel("$y$ = Actual SalePrice")
    plt.legend(loc = "upper left")
    plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], c='red')
    plt.tight_layout()
```

```python
def exponentiate(vector):
    return(pd.Series(vector).apply(lambda num: np.exp(num)))
```

### Determining Baseline Score

The baseline score is the $R^2$ score of a model that predicts the mean housing price for all observations in the sample.

```python
# Instantiate: creates a dummy regression that always predicts the mean of the target
base_mean = DummyRegressor(strategy='mean')

# Fit the "model"
base_mean = base_mean.fit(X_train, y_train)

print("Training Set R^2 Score is:", base_mean.score(X_train, y_train))
print("Testing Set R^2 Score is:",  base_mean.score(X_test, y_test))
```

### Linear Regression

```python
# Instantiate and fit
lr = LinearRegression()
ols = lr.fit(X_train, y_train)

# Predict y_hat
yhat_train_OLS = ols.predict(X_train)
yhat_test_OLS = ols.predict(X_test)
```

```python
# Score it
ols_cvs = cross_val_score(ols, X_train, y_train, cv=5).mean()
print("K=5 Fold Cross Validation Score is:", np.round(ols_cvs,3)) 
print("Training Set R^2 Score is:", np.round(ols.score(X_train, y_train),3))
print("Testing Set R^2 Score is:", np.round(ols.score(X_test, y_test),3))
```

```python
# Plot it
plot_predictions(model_name = "Linear ",
                 y_train_pred = exponentiate(yhat_train_OLS),
                 y_test_pred = exponentiate(yhat_test_OLS),
                 y_train_actual = y_train,
                 y_test_actual = y_test)
```

```python
# Visualize coefficients
coefs = pd.Series(ols.coef_, index = X_train.columns)
large_coefs = pd.concat([coefs.sort_values().head(10), coefs.sort_values().tail(10)])
large_coefs.plot(kind = "barh")
plt.title("Coefficients in the OLS Linear Regression Model")
plt.show()
```

Clearly the coefficients are blown out of proportion given that they are in magnitudes of $10^{16}$! This is expected since we have over 500 regressors and only a little over 500 observations in the testing set and cross-validation subsets. The Training Set had 2049 observations so we saw a reasonable R^2 there. When the number of regressors approaches the number of observations, coefficient estimates begin to explode. This is precisely why we need regularization models.

We consider Ridge and Lasso models below:

### Ridge Regression

```python
# Instantiate and fit, while tuning hyperparametrs: optimizing alpha
parameter_grid = {'alpha': [0.01, 0.1, 1, 5, 10, 25, 50, 100]}
ridge = GridSearchCV(Ridge(), cv=5, param_grid=parameter_grid, scoring='r2')
ridge.fit(X_train, np.log(y_train))
alpha = ridge.best_params_['alpha']

# Re-optimize alpha based on another grid around this optimal alpha
parameter_grid = {'alpha': [x/100. * alpha for x in range(50, 150, 5)]}
ridge = GridSearchCV(Ridge(), cv=5, param_grid=parameter_grid, scoring='r2')
ridge.fit(X_train, np.log(y_train))
alpha = ridge.best_params_['alpha']
ridge = ridge.best_estimator_
print("Optimal Alpha:",alpha)
```

```python
# Predict y_hat
yhat_train_ridge = ridge.predict(X_train)
yhat_test_ridge = ridge.predict(X_test)
```

```python
# Score it
ridge_cvs = cross_val_score(ridge, X_train, np.log(y_train), cv=5).mean()
print("K=5 Fold Cross Validation Score is:", np.round(ridge_cvs,3)) 
print("Training Set R^2 Score is:", np.round(ridge.score(X_train, np.log(y_train)),3))
print("Testing Set R^2 Score is:", np.round(ridge.score(X_test, np.log(y_test)),3))
```

It's unsurprising that the Training Set $R^2$ score was the highest out of the three metrics, but with $R^2$ scores of more than 0.90 across all 3 cases, especially on the testing set, our Ridge Model does an excellent job of capturing the variance in actual `SalePrice`. However, there is a non-trivial 4.2% difference between the Training and Testing scores which suggests that the model is suffering from overfitting.

```python
# Plot it
plot_predictions(model_name = "Ridge ",
                 y_train_pred = exponentiate(yhat_train_ridge),
                 y_test_pred = exponentiate(yhat_test_ridge),
                 y_train_actual = y_train,
                 y_test_actual = y_test)
```

After plotting our predicted values against the actual values, it seems that the Ridge Regression does a very good job predicting prices of homes under 500,000 USD. However, our model does not do as well predicting the price of homes $>$ 500,000 USD though fortunately there are not many cases of homes of prices this high.

### Lasso Regression

```python
# Instantiate and fit, while tuning hyperparameters: optimizing alpha
parameter_grid = {'alpha': list(np.arange(0.1, 1.5, 0.1)),
                  'tol': [0.001]}
lasso = GridSearchCV(Lasso(), cv=5, param_grid=parameter_grid, scoring='r2')
lasso.fit(X_train, np.log(y_train))
alpha = lasso.best_params_['alpha']
Lasso()

# Re-optimize alpha
parameter_grid = {'alpha': [(x/100)*alpha for x in range(50, 150, 5)],
                  'tol': [0.001]}
lasso = GridSearchCV(Lasso(), cv=5, param_grid=parameter_grid, scoring='r2')
lasso.fit(X_train, np.log(y_train))
alpha = lasso.best_params_['alpha']
lasso = lasso.best_estimator_
```
